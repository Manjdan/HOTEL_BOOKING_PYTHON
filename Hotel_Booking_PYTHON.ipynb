```python
def info_of_cat():
    for col in list_1 : 
        print(f"Unique value of {col}:{df[col].unique()}")
        print(f"mode vale of {col}:{df[col].mode()}")
        print(f"Number of missing values in {col} is {df[col].isnull().sum()}")
        print("---------------------------------")
```

```python
def info_of_numerical():
    for col in list_2:
        print(f"The mean of the {col} is {df[col].mean()}")
        print(f"The median of the {col} is {df[col].median()}")
        print(f"The mode of the {col} is {df[col].mode()[0]}")
        print(f"The standard deviation of the {col} is {df[col].std()}")
        print(f"Number of missing values in the {col} is {df[col].isnull().sum()}")
        print("--------------------------------------------------------------------\n")
info_of_numerical()
```

```python
list_2 = df.select_dtypes(exclude= "object")
def outliers(column):
    for column in list_2:
        print(f'The mean of the {column} is , {df[column].mean()}')
        Q1=df[column].quantile(0.25)
        Q3=df[column].quantile(0.75)
        print('value of Q1 and Q3',Q1,',',Q3)
        IQR=Q3-Q1
        print('value of IQR',IQR)
        UF = Q3 + (1.5 * IQR) 
        LF = Q1 - (1.5 * IQR)
        print('value of UF and LF',UF,',',LF)
        outlier=[]
        for i in df[column]: 
                if (i<LF or i>UF):
                    outlier.append(i)

        print("Outliers from IQR method: ",len(outlier))
        print('#'*30)
```

• What are the peak booking periods for the hotel?


```python
def bar_percentage(columns,df):
    fig,ax = plt.subplots(figsize = (3,3))
    sns.countplot(x = columns ,hue='hotel',data = df , ax = ax)
    plt.xticks(rotation=30)
    for bar in ax.patches:
        percentage = f"{round((bar.get_height()/df.shape[0]) *100,2)}%"
        x = bar.get_x()+ bar.get_width()/2
        y = bar.get_height()
        ax.annotate(percentage,(x,y),va = "bottom",ha = "center")
    plt.show()
    
bar_percentage("arrival_date_year",df)
```
